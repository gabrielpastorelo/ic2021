```
import pandas as pd

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/gabarito.csv'
gabarito = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/prefci.csv'
pre_fci = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/posfci.csv'
pos_fci = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/quest1.csv'
quest1 = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/quest2.csv'
quest2 = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/quest3.csv'
quest3 = pd.read_csv(url)

url = 'https://raw.githubusercontent.com/gabrielpastorelo/ic2021/master/quest4.csv'
quest4 = pd.read_csv(url)
```

```
matrix = {'index' : [*range(0,len(pre_fci))],
          'tempo_prefci' : [None]*len(pre_fci),
          'alternativas_prefci' : [None]*len(pre_fci),
          'tempo_posfci' : [None]*len(pre_fci),
          'alternativas_posfci' : [None]*len(pre_fci),
          'tempo_quest1' : [None]*len(pre_fci),
          'nota_quest1' : [None]*len(pre_fci),
          'tempo_quest2' : [None]*len(pre_fci),
          'nota_quest2' : [None]*len(pre_fci),
          'tempo_quest3' : [None]*len(pre_fci),
          'nota_quest3' : [None]*len(pre_fci),
          'tempo_quest4' : [None]*len(pre_fci),
          'nota_quest4' : [None]*len(pre_fci)
          }

for i in pre_fci['Index']:
  matrix['tempo_prefci'][i-1] = pre_fci['Tempo'][i-1]
  alternativas = []
  for j in range(3,33):
    alternativas.append(pre_fci.loc[i-1][j])
  matrix['alternativas_prefci'][i-1] = alternativas

k = 0
for i in pos_fci['Index']:
  matrix['tempo_posfci'][i-1] = pos_fci['Tempo'][k]
  alternativas = []
  for j in range(3,33):
    alternativas.append(pos_fci.loc[k][j])
  matrix['alternativas_posfci'][i-1] = alternativas
  k += 1

k = 0
for i in quest1['Index']:
  if matrix['tempo_quest1'][i-1] == None:
    matrix['tempo_quest1'][i-1] = [quest1['Tempo'][k]]
    matrix['nota_quest1'][i-1] = [quest1['Nota'][k]]
  else:
    matrix['tempo_quest1'][i-1].append(quest1['Tempo'][k])
    matrix['nota_quest1'][i-1].append(quest1['Nota'][k])
  k += 1

k = 0
for i in quest2['Index']:
  if matrix['tempo_quest2'][i-1] == None:
    matrix['tempo_quest2'][i-1] = [quest2['Tempo'][k]]
    matrix['nota_quest2'][i-1] = [quest2['Nota'][k]]
  else:
    matrix['tempo_quest2'][i-1].append(quest2['Tempo'][k])
    matrix['nota_quest2'][i-1].append(quest2['Nota'][k])
  k += 1

k = 0
for i in quest3['Index']:
  if matrix['tempo_quest3'][i-1] == None:
    matrix['tempo_quest3'][i-1] = [quest3['Tempo'][k]]
    matrix['nota_quest3'][i-1] = [quest3['Nota'][k]]
  else:
    matrix['tempo_quest3'][i-1].append(quest3['Tempo'][k])
    matrix['nota_quest3'][i-1].append(quest3['Nota'][k])
  k += 1

k = 0
for i in quest4['Index']:
  if matrix['tempo_quest4'][i-1] == None:
    matrix['tempo_quest4'][i-1] = [quest4['Tempo'][k]]
    matrix['nota_quest4'][i-1] = [quest4['Nota'][k]]
  else:
    matrix['tempo_quest4'][i-1].append(quest4['Tempo'][k])
    matrix['nota_quest4'][i-1].append(quest4['Nota'][k])
  k += 1
```

```
df = pd.DataFrame(data = matrix)
```
